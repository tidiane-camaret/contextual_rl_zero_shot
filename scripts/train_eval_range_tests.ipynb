```python
import numpy as np
x = np.arange(10)
y = np.geomspace(0.01, 10, 10)

#plot the data
import matplotlib.pyplot as plt
plt.plot(x, y)  
# plot y = 0.02
plt.axhline(y=1, color='r', linestyle='--')
plt.show()
```
